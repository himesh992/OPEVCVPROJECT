```python
# Step 2: Import the libraries
from paddleocr import PaddleOCR
from google.colab import files
import os
import matplotlib.pyplot as plt
import cv2

# Step 3: Upload image(s) from your local system
uploaded = files.upload()

# Verify the uploaded file
for filename in uploaded.keys():
    print(f"Uploaded file: {filename}")

# Get the image path
image_path = "/content/" + list(uploaded.keys())[0]  # Fetch the filename

# Step 4: Initialize PaddleOCR
ocr = PaddleOCR(use_angle_cls=True, lang='en')  # 'en' for English language

# Step 5: Run OCR on the uploaded image
result = ocr.ocr(image_path, cls=True)

# Step 6: Extract and print the results
extracted_text = ""
for line in result[0]:
    extracted_text += line[1][0] + "\n"

# Print the extracted text
print(f"Extracted Text using PaddleOCR:\n{extracted_text}")

# Step 7: Display the image
image = cv2.imread(image_path)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for proper display
plt.axis('off')  # Hide the axes for better visualization
plt.title("Uploaded Image with Extracted Text")
plt.show()
```
